```python
import cv2
import pytesseract

# Load the image
image_path = 'D:/STUDY/MS/Amazon ML Hackathon 2024/student_resource 3/images/41nblnEkJ3L.jpg'
image = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding
thresh = cv2.adaptiveThreshold(
    gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2
)

# Remove noise using median blurring
# blur = cv2.medianBlur(thresh, 3)

# Optionally, resize the image for better accuracy (if needed)
resize = cv2.resize(thresh, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)

# Apply Tesseract to the processed image
pytesseract.pytesseract.tesseract_cmd = r'C:/Program Files/Tesseract-OCR/tesseract.exe'
text2 = pytesseract.image_to_string(resize)

print(text2)
```
